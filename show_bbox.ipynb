```python
import pandas as pd
df = pd.read_csv('./yolo7_5236.csv')
df
```

```python
import cv2
import matplotlib.pyplot as plt
import os

classes = {0:"general trash",1:"Paper",2:"Paper pack",3:"Metal",4:"Glass",5:"Plastic",6:"Styrofoam",7:"Plastic bag",8:"Battery",9:"Clothing"}

# 결과 이미지를 저장할 디렉토리 생성
output_dir = 'output_bbox_images'
os.makedirs(output_dir, exist_ok=True)

# 전체 데이터프레임을 순회
for i in range(len(df)):
    string = str(df.loc[i, 'PredictionString'])
    image_id = '/data/ephemeral/home/data/dataset/test/' + df.loc[i, 'image_id'][-8:]
    image = cv2.imread(image_id)
    
    if image is None:
        print(f"Unable to read image: {image_id}")
        continue

    string = string.split(" ")
    obj = []
    
    for j in range(0, len(string), 6):
        obj.append(string[j:j+6])
        
    for k in obj:
        label_num = int(k[0])  # 첫 번째 요소를 라벨로 사용
        label = classes[label_num]
        confidence = float(k[1])  # 두 번째 요소를 신뢰도로 사용
        pt1 = (int(float(k[2])), int(float(k[3])))
        pt2 = (int(float(k[4])), int(float(k[5])))
        
        # 바운딩 박스 그리기
        cv2.rectangle(image, pt1, pt2, (0, 255, 0), 2)
        
        # 라벨 텍스트 준비
        label_text = f"{label}: {confidence:.2f}"
        
        # 라벨 배경 크기 계산
        text_size = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)[0]
        
        # 라벨 배경 그리기
        cv2.rectangle(image, 
                      (pt1[0], pt1[1] - text_size[1] - 4), 
                      (pt1[0] + text_size[0], pt1[1]), 
                      (0, 255, 0), 
                      -1)
        
        # 라벨 텍스트 그리기
        cv2.putText(image, label_text, 
                    (pt1[0], pt1[1] - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)

    # 이미지 저장 (필요한 경우 주석 해제)
    output_path = os.path.join(output_dir, f'output_{i}.png')
    cv2.imwrite(output_path, image)
    
    # 진행 상황 출력 (100개마다)
    if (i + 1) % 100 == 0:
        print(f"Processed {i + 1} images")

    # 처리된 이미지 표시 (필요한 경우 주석 해제)
    # plt.figure(figsize=(10, 10))
    # plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    # plt.axis('off')
    # plt.show()

print("All images processed")
```
